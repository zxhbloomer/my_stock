# 数据探索

参考: qlib/examples/tutorial/detailed_workflow.ipynb

```python
import qlib
from qlib.constant import REG_CN
from qlib.data import D
import pandas as pd
```

## 初始化 Qlib

```python
qlib.init(provider_uri="~/.qlib/qlib_data/cn_data", region=REG_CN)
```

## 查看股票池

```python
# 获取 CSI300 成分股
instruments = D.instruments(market='csi300')
print(f"CSI300 成分股数量: {len(instruments)}")
print(f"示例: {instruments[:5]}")
```

## 获取股票数据

```python
# 获取贵州茅台的数据
df = D.features(
    ["SH600519"],
    ["$open", "$high", "$low", "$close", "$volume"],
    start_time="2020-01-01",
    end_time="2020-12-31",
)
df.head(10)
```

## 查看市场整体情况

```python
# 统计每日有数据的股票数量
df_all = D.features(
    D.instruments("csi300"),
    ["$close"],
    start_time="2020-01-01",
    end_time="2020-12-31"
)

daily_count = df_all.groupby("datetime").size()
print(f"平均每日股票数量: {daily_count.mean():.0f}")
print(f"最少: {daily_count.min()}")
print(f"最多: {daily_count.max()}")
```
